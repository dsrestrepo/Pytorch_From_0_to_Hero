```python
import torch
import torchvision
import torchvision.transforms as transforms
```

```python
# See the shape of all images:
# Main Directory
main_path = 'Data/flowers/'

import os
from PIL import Image
import numpy as np

def get_size(folder):
    sizes = []
    # Get from main directory all sub-directories
    for folder_path in os.listdir(folder):
        # Skip
        if (folder_path == '.DS_Store'):
            continue
        # in all sub Directories take the dimesions of each image
        for filename in os.listdir(os.path.join(folder, folder_path)):
            # take image
            img = Image.open(os.path.join(folder, folder_path, filename))
            # Get image with 
            #print(img.size)
            # Stores data like: (width, height)
            sizes.append(img.size)
            
    print(f'the max width is: {max(sizes[0])}, and the min width is: {min(sizes[0])}')
    print(f'the max height is: {max(sizes[1])}, and the min height is: {min(sizes[1])}')
    print(f'the mean width is: {np.mean(sizes[0])}, and the mean height is: {np.mean(sizes[1])}')
    
# Call the function
get_size(main_path)
```

```python
#Transforms
transform_train = transforms.Compose([
    transforms.Resize((200,200)),
    #transforms.RandomResizedCrop(199),
    transforms.RandomRotation(30),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),  #0-255 to 0-1, numpy to tensors
    transforms.Normalize([0.5,0.5,0.5], # 0-1 to [-1,1] , formula (x-mean)/std
                        [0.5,0.5,0.5])
])

transform_test = transforms.Compose([transforms.Resize((200, 200)),
                                         transforms.ToTensor(),
                                         transforms.Normalize([0.5,0.5,0.5], # 0-1 to [-1,1] , formula (x-mean)/std
                                                             [0.5,0.5,0.5])
])

transform_validation = transforms.Compose([transforms.Resize((200, 200)),
                                         transforms.ToTensor(),
                                         transforms.Normalize([0.5,0.5,0.5], # 0-1 to [-1,1] , formula (x-mean)/std
                                                             [0.5,0.5,0.5])
])
```

```python
# Read Dataset
# Main folder 
main_dir = 'Data/flowers_split/'
# Read image in structure main_folder -> subfolders -> images
dataset_train = torchvision.datasets.ImageFolder(os.path.join(main_dir, 'train'), transform=transform_train)
dataset_test = torchvision.datasets.ImageFolder(os.path.join(main_dir, 'test'), transform=transform_test)
dataset_validation = torchvision.datasets.ImageFolder(os.path.join(main_dir, 'val'), transform=transform_validation)

# If I just have one dir:
# dataset = torchvision.datasets.ImageFolder(main_dir, transform=transform_test)
# Take 85% train 15% Test
# train = (len(dataset) * 85) // 100
# test = len(dataset) - train

# Split in Train and Test
# train_set, test_set = torch.utils.data.random_split(dataset, [train, test])

#print(dataset.class_to_idx)
#print(f'The classes are; {dataset.classes}')
# See all labels
#print(dataset.targets)
# See Path to image and class
#for i in dataset.samples:
#    print(i)

```

```python
# Dataloader
# Using the image datasets and the trainforms, define the dataloaders
train_loader = torch.utils.data.DataLoader(dataset_train, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset_test, batch_size=32)
val_loader = torch.utils.data.DataLoader(dataset_validation, batch_size=32)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# functions to show an image
def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(train_loader)
images, labels = dataiter.next()

# show images
imshow(torchvision.utils.make_grid(images))
```

```python
# Calculo 
# output_width = ((W-F+2*P )/S)+1
# W = width of the image
# F = Filter size
# P = Padding (pixels at sides)
# S = Stride
output_width = ((200-5+2*0)/1)+1
print(output_width)
# Out Dim = 6
output_width = ((output_width-2+2*0)/2)+1
print(output_width)
# Out Dim = 16
output_width = ((output_width-5+2*0)/1)+1
print(output_width)
# Out Dim = 16
output_width = ((output_width-2+2*0)/2)+1
print(output_width)
print(output_width*output_width*16)
```

```python
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.conv1 = nn.Conv2d(3, 6, 5)
        # Out Dim = 6
        output_width = ((200-5+2*0)/1)+1
        self.pool = nn.MaxPool2d(2, 2)
        # Out Dim = 6
        output_width = ((output_width-2+2*0)/2)+1
        self.conv2 = nn.Conv2d(6, 16, 5)
        # Out Dim = 16
        output_width = int(((output_width-5+2*0)/1)+1)
        # Out Dim = 16
        output_width = int(((output_width-2+2*0)/2)+1)
        self.fc1 = nn.Linear(16 * output_width * output_width, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 5)
        #self.out = nn.LogSoftmax(dim=1)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        #x = self.out(x)
        return x


net = Net()
```

```python
import torch.optim as optim
import numpy as np

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)

for epoch in range(70):  # loop over the dataset multiple times

    running_loss = []
    steps = 0
    
    for inputs, labels in iter(train_loader):
        steps += 1
        #inputs, labels = inputs.to('cuda'), labels.to('cuda')
        net.train()

        # zero the parameter gradients
        optimizer.zero_grad()
        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        # add loss to array
        running_loss.append(loss.item())
        
        # Do validation each 20 steps in epoch % 10
        if ((steps % 20 == 0) and (epoch % 5 == 0)):
            
            net.eval()
            # Turn off gradients for validation, saves memory and computations
            with torch.no_grad():
                val_loss = []
                for val_images, val_labels in iter(val_loader):
                    #val_images, val_labels = val_images.to('cuda'), val_labels.to('cuda')
                    output_val = net.forward(val_images)
                    val_loss.append(criterion(output_val, val_labels).item())

            print(f'Epoch {epoch} - Step: {steps}: validation loss: {np.mean(val_loss)} - train loss: {np.mean(running_loss)}')
            
    if epoch % 10 == 0:    # print every 10 epoch
        print(f'The loss are {np.mean(running_loss)}')
print('Finished Training')

# Save the model
#PATH = './flowers_cnn.pth'
#torch.save(net.state_dict(), PATH)
```

```python
# Save the model
PATH = './flowers_cnn.pth'
torch.save(net.state_dict(), PATH)
```

```python
# Do validation on the test set
net.eval()

correct = 0
total = 0
# since we're not training, we don't need to calculate the gradients for our outputs
with torch.no_grad():
    for images, labels in iter(test_loader):
        # calculate outputs by running images through the network
        outputs = net(images)
        #print(labels)
        # the class with the highest energy is what we choose as prediction
        _, predicted = torch.max(outputs.data, 1)
        #print(predicted)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Accuracy of the network on the test images: {(100 * correct / total)}' )
```
